```python
# --- Smoke test for app artefacts before deploy ---
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

from pathlib import Path
import pandas as pd

ARTE_ROOT = Path("/content/drive/MyDrive/gt-markets/AppDemo/artefacts")
ASSETS=["GOLD","BTC","OIL","USDCNY"]
FREQS=["D","W"]

rows=[]
for a in ASSETS:
    for F in FREQS:
        asset_dir=ARTE_ROOT/a
        rows.append({
            "asset":a,"freq":F,
            "metrics_baseline": "✓" if (asset_dir/f"metrics_baseline_{F}.csv").exists() else "✗",
            "metrics_keywords": "✓" if (asset_dir/f"metrics_keywords_{F}.csv").exists() else "✗",
            "any_signals": "✓" if list(asset_dir.glob(f"signals_*_{F}.csv")) else "✗",
            "any_figs": "✓" if list((asset_dir/"figs").glob(f"*_{F}.png")) else "✗",
        })
df=pd.DataFrame(rows)
print("Artefacts root:", ARTE_ROOT)
df
```
