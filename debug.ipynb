```python
# run this from the repo root, e.g. /content/gt-markets or your local clone
from pathlib import Path
import pandas as pd

ARTE = Path("AppDemo/artefacts")
assets = ["GOLD","BTC","OIL","USDCNY"]
freqs  = ["D","W"]

def want_strats(asset_dir, F):
    want = set()
    for name in [f"metrics_baseline_{F}.csv", f"metrics_keywords_{F}.csv"]:
        p = asset_dir / name
        if p.exists():
            try:
                df = pd.read_csv(p)
                if "strategy" in df.columns:
                    want |= set(df["strategy"].dropna().astype(str).unique())
            except Exception:
                pass
    return sorted(want)

missing = []
for a in assets:
    a_dir = ARTE / a
    for F in freqs:
        # required metrics
        for fn in [f"metrics_baseline_{F}.csv"]:
            if not (a_dir/fn).exists():
                missing.append(str(a_dir/fn))
        # optional keywords metrics (only if you expect keyword strategies to appear)
        # if not (a_dir/f"metrics_keywords_{F}.csv").exists():
        #     missing.append(str(a_dir/f"metrics_keywords_{F}.csv"))

        # required signals for every strategy listed in metrics_*.csv
        for s in want_strats(a_dir, F):
            sig = a_dir / f"signals_{s}_{F}.csv"
            if not sig.exists():
                missing.append(str(sig))

print("MISSING:\n" + ("\n".join(missing) if missing else "Nothing ✨"))
```
